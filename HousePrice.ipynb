```python
# function to create histogram, QQ plot and boxplot

# for Q-Q plot
import scipy.stats as stats
 
def diagnostic_plots(df, variable):
    # function takes a dataframe (df) and
    # the variable of interest as arguments
 
    # define figure size
    plt.figure(figsize=(16, 4))
 
    # histogram
    plt.subplot(1, 3, 1)
    sns.distplot(df[variable], bins=30)
    plt.title('Histogram')
 
    # Q-Q plot
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.ylabel('Variable quantiles')
 
    # boxplot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')
 
    plt.show()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)

    # Calculate and print the percentage of data retained
    data_retained_percentage = (len(df_out) / len(df)) * 100
    print(f"Data retained after outlier removal: {data_retained_percentage:.2f}%")

    return df_out
```

```python
def plot_scatter_chart(df,location):
  bhk2 = df[(df.location==location) & (df.bhk==2)]
  bhk3 = df[(df.location==location) & (df.bhk==3)]
  plt.figure(figsize=(16,9))
  plt.scatter(bhk2.total_sqft_int, bhk2.price, color='Blue', label='2 BHK', s=50)
  plt.scatter(bhk3.total_sqft_int, bhk3.price, color='Red', label='3 BHK', s=50, marker="+")
  plt.xlabel("Total Square Feet Area")
  plt.ylabel("Price")
  plt.title(location)
  plt.legend()
 
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')
```

```python
def update_availability(row):
    # Leave "Ready to Move" as it is
    if row == 'Ready To Move':
        return 'Ready To Move'

    # Extract the month part from the original availability
    original_parts = row.split('-')

    # Check if the second part is a numeric day, indicating a specific date
    if len(original_parts) == 2 and original_parts[1].isalpha():
        return original_parts[1]

    return row

# Apply the mapping to the 'Availability' column
df11['availability'] = df11['availability'].apply(update_availability)
```

```python
from datetime import datetime

# Function to dynamically update the availability column
def update_availability(row):
    current_month = datetime.now().strftime('%b')

    if row == 'Ready To Move':
        return 'Ready To Move'

    # Update the availability based on the current month
    return 'This Month' if row == current_month else row

# Apply the mapping to the 'availability' column
df11['availability'] = df11['availability'].apply(update_availability)
```

```python
# Train
from sklearn.metrics import mean_squared_error
#calculates the root mean squared error (RMSE) between predicted values (y_pred) and actual values (y_test). 

def rmse(y_test, y_pred):
  return np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
rfr_score=rfr.score(X_test,y_test) # The rfr.score() method returns the coefficient of determination R^2 of the prediction.
# indicates that your model explains approximately 87.04% of the variance in the target variable. 
rfr_rmse = rmse(y_test, rfr.predict(X_test)) # represents the average magnitude of the errors in your predictions. 
rfr_mse = mean_squared_error(y_test,rfr.predict(X_test)) # is the average squared difference between the predicted and actual values. 
rfr_score, rfr_rmse,rfr_mse
```

```python
import xgboost
xgb_reg = xgboost.XGBRegressor()
xgb_reg.fit(X_train,y_train)
xgb_reg_score=xgb_reg.score(X_test,y_test) 
xgb_reg_rmse = rmse(y_test, xgb_reg.predict(X_test))
xgb_reg_mse = mean_squared_error(y_test,xgb_reg.predict(X_test))
xgb_reg_score, xgb_reg_rmse,xgb_reg_mse
```

```python
# Predictions from RandomForestRegressor and XGBoostRegressor
from sklearn.metrics import r2_score
rfr_pred = rfr.predict(X_test)
xgb_pred = xgb_reg.predict(X_test)

# Combined prediction by averaging
combined_pred = (rfr_pred + xgb_pred) / 2

# Calculate R^2 score for combined prediction
combined_r2_score = r2_score(y_test,combined_pred)

# Calculate RMSE for combined prediction
combined_rmse = rmse(y_test,combined_pred)

# calculate mse
combined_mse = mean_squared_error(y_test,combined_pred)

combined_r2_score, combined_rmse, combined_mse
```

```python
def predict_house_price(bath, balcony, total_sqft_int, bhk, availability, location):
    x = np.zeros(len(X.columns))  # Assuming X is your feature matrix
    current_month = datetime.now().strftime('%b')
    
    # Assigning feature values to respective positions in the array
    x[0] = bath
    x[1] = balcony
    x[2] = total_sqft_int
    x[3] = bhk
  

    if availability == "Ready To Move":
        x[7] = 1
    
    # Dynamically set the current month feature
    month_feature_index = np.where(X.columns == 'availability_' + current_month)[0]
    if month_feature_index:
        x[month_feature_index[0]] = 1


    if 'location_' + location in X.columns:
        loc_index = np.where(X.columns == 'location_' + location)[0][0]
        x[loc_index] = 1

    # Feature scaling
    x = sc.transform([x])[0]  # Assuming 'sc' is a fitted scaler

    # Predictions from pre-fixed models (loaded_xgb_reg and loaded_rfr)
    prediction1 = loaded_xgb_reg.predict([x])[0]  # Replace loaded_xgb_reg with your XGBoost model
    prediction2 = loaded_rfr.predict([x])[0]     # Replace loaded_rfr with your RandomForest model

    # Combine predictions (averaging)
    combined_prediction = (prediction1 + prediction2) / 2

    return combined_prediction
```
